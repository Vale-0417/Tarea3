```python
promedio_menor_igual_129 = Datos.loc[Datos['trestbps'] <= 129, 'trestbps'].mean()

Datos.loc[Datos['trestbps'] > 129, 'trestbps'] = promedio_menor_igual_129

print("Cantidad de pacientes que cuentan con la presión más alta que 129 corregidos por el promedio menor o igual a 129:", (Datos['trestbps'] > 129).sum())
```

```python
promedio_menor_igual_190 = Datos.loc[Datos['chol'] <= 190, 'chol'].mean()

# Reemplazar los valores mayores que 190 por el promedio de los valores menores o iguales a 190
Datos.loc[Datos['chol'] > 190, 'chol'] = promedio_menor_igual_190

print("Cantidad de pacientes que cuentan con el colesterol más alto que 190 corregidos por el promedio menor o igual a 190:", (Datos['chol'] > 190).sum())
```
